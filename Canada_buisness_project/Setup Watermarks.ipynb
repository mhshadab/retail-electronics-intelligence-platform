```python
# Databricks notebook source
# MAGIC %md
# MAGIC # 00 — Setup Watermark Table (Run Once)
# MAGIC Creates `canada_business.bronze.ingest_watermarks` to track incremental state.

# COMMAND ----------

spark.sql("""
CREATE TABLE IF NOT EXISTS canada_business.bronze.ingest_watermarks (
    source        STRING    COMMENT 'Source system: fred, google_trends, wits, statcan',
    series_key    STRING    COMMENT 'Series ID (FRED), keyword (Trends), PID (StatCan), or ALL (WITS)',
    last_obs_date DATE      COMMENT 'Max observation/ref_date ingested for this key',
    last_run_ts   STRING    COMMENT 'Timestamp of last successful ingest run (YYYYMMDDTHHMMSSz)',
    rows_written  LONG      COMMENT 'Rows written in the last run',
    updated_at    TIMESTAMP COMMENT 'When this watermark row was last updated'
)
USING DELTA
COMMENT 'Incremental ingest watermarks — tracks what has been ingested per source/series'
""")

print("✅ Watermark table created: canada_business.bronze.ingest_watermarks")

# COMMAND ----------

# Verify
display(spark.sql("SELECT * FROM canada_business.bronze.ingest_watermarks"))




























```
